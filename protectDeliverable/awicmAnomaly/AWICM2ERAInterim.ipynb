```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import xesmf as xe
```

```python
path='/isibhv/netscratch/orichter/AWICM/ssp370_2090-2099_historical_1990-1999_diff.nc'
ds=xr.open_dataset(path)
path='/isibhv/netscratch/orichter/AWICM/eraint_years/era_1979/dlwrf_00_12_1979.nc'
ds_out=xr.open_dataset(path)
```

```python
ds_out
```

# Interpolation

```python
regridder=xe.Regridder(ds,ds_out,'bilinear',periodic=True)
regridder
```

```python
ds_out=regridder(ds)
ds_out
```

```python
ds_out=ds_out.squeeze().drop('time')
```

# Conversion

```python
#radiation, convert Watts/m2 to Joule/m2 in 12h
ds_out['ssrd']=ds_out.DSWR*3600*12
ds_out['strd']=ds_out.DLWR*3600*12
```

```python
#evaporation and precipitation, convert kg/m2/s to m in 12h
rho_w = 1000 #kg/m3
ds_out['e']=ds_out.evap*3600*12/rho_w
ds_out['tp']=ds_out.PREC*3600*12/rho_w
```

```python
#temperature
ds_out['t2m']=ds_out.temp2
```

```python
#dewPoint, convert specific humidity (kg/kg) and temperature (K) in dew point temperature (K)
# formular is apparently Td = T - ((100 - RH)/5. if RG in %
# I'm not sure if this is okay, but I'm out of time
ds_out['d2m']=ds_out.temp2-((1-ds_out.q2m)/5)
```
