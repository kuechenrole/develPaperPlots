```python
import sys
import os
sys.path.append("../pyfesom")

import pyfesom as pf
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import LinearSegmentedColormap, from_levels_and_colors
import numpy as np
#%matplotlib notebook
%matplotlib inline
from matplotlib import cm
from netCDF4 import Dataset, MFDataset
import xarray as xr
import pdb
import cmocean.cm as ocm
import glob
import pandas as pd
import matplotlib
```

```python
masterPath='/albedo/work/projects/oce_rio/orichter/ollieWork/Misomip2/'
def get_mesh(exp,cavity=False):
    meshpath  =os.path.join(masterPath,exp,'fesomInitialMesh')
    mesh = pf.fesom_mesh(meshpath, cavity=cavity)
    
    return mesh

def get_frc(exp,year):
    dataPath=os.path.join(masterPath,exp,'fesomData','%s.%s.forcing.diag.nc'%(exp,year))
    return xr.open_dataset(dataPath)

def get_ax(ax,data,mesh,levels='max',cmap=ocm.balance,title='',cbarlabel=""):

    level_data, elem_no_nan = pf.get_data(data,mesh,0)
        
    if levels=='max':
        levels=np.linspace(level_data.min(),level_data.max(),100)
    else:
        levels = levels  
    level_data[level_data>max(levels)]=max(levels);
    level_data[level_data<min(levels)]=min(levels);
    
    #m=Basemap(projection='spstere',boundinglat=-60,lon_0=180)
    #x, y = m(mesh.x2, mesh.y2)
    #m.drawmapboundary(fill_color='0.9',ax=ax)
    wpl = ax.tricontourf(mesh.x2,mesh.y2, elem_no_nan[::], level_data, levels = levels,
                    cmap=cmap, extend='both')
    
    cbar = plt.colorbar(wpl,ax=ax,orientation='vertical', pad=0.03);
    cbar.set_label(cbarlabel)
    ax.set_title(title)
    #ax.set_ylabel('Latitude ($\circ$C)')
    #ax.set_xlabel('Longitude ($\circ$C)')
    #ax.axis_ticks('off')
    plt.tight_layout()
    
    return ax,levels
```

```python
def comp(key,same_cscale=True):
    fig,axes = plt.subplots(ncols=2,figsize=(12,5))
    ax1,ax2=axes.flatten()
    ax1,levels=get_ax(ax1,frc20[key][1],mesh20,title='%s; Era-Interim'%(frc20[key].description),cbarlabel=frc20[key].units)
    if same_cscale:
        ax2,levels=get_ax(ax2,frc25[key][1],mesh25,title='%s; new'%(frc25[key].description),cbarlabel=frc25[key].units,
                     levels=levels)
    else:
        ax2,levels=get_ax(ax2,frc25[key][1],mesh25,title='%s; new'%(frc25[key].description),cbarlabel=frc25[key].units)   
    plt.show()
    
for key in frc20: 
#['rain','snow','runoff','swrd','qnet','olat','osen','olwout','wnet','virtual_salt','relax_salt',
#            'stress_x','stress_y','Tsurf','Ssurf']:
    if frc20[key].ndim<2: continue
    if frc20[key][1].sum()==0: continue
    print(key)
    comp(key)
```

```python
def plot_melt(ax,exp,fileHandle,region=False,levels='max',figsize=(9,11),ollie=False):
    
    #masterPath='/work/ollie/orichter/Misomip2/'
    if ollie:
        masterPath='/albedo/work/projects/oce_rio/orichter/ollieWork/Misomip2/'
        meshpath  =os.path.join(masterPath,exp,'fesomInitialMesh')
        dataPath=os.path.join(masterPath,exp,'fesomData','%s.%s'%(exp,fileHandle))
    else:
        masterPath='/isibhv/netscratch/orichter/'
        meshpath  =os.path.join(masterPath,'timmsmesh',exp)
        dataPath=os.path.join(masterPath,'timmsdata',exp,'%s.%s'%(exp,fileHandle))
   
    mesh = pf.fesom_mesh(meshpath, cavity=False,get3d=True)
    frc = xr.open_mfdataset(dataPath)
    wnet = frc.wnet.mean('T').where(mesh.cflag).load()*365.25*24*3600
    level_data, elem_no_nan = pf.get_data(wnet,mesh,0)
    #pdb.set_trace()

    
    #plt.figure(figsize=figsize)
    if region==False:
        m=Basemap(projection='spstere',boundinglat=-64,lon_0=180)
        m.drawmapboundary(fill_color='0.9')
        x, y = m(mesh.x2, mesh.y2)
        m.bluemarble(alpha=0.5)
    else:
        x, y = mesh.x2, mesh.y2
        ax.set_xlim([region[0],region[1]])
        ax.set_ylim([region[2],region[3]])

    if levels=='max':
        levels=np.arange(np.nanmin(level_data),np.nanmax(level_data))
    else:
        levels = levels

    mp = ax.tricontourf(x, y, elem_no_nan[::], level_data, levels = levels, cmap=ocm.thermal, extend='both')
    #cbar = plt.colorbar(orientation='horizontal');
    #cbar.set_label("ice shelf melt rate, m/yr")
    #plt.title('%s, %s mean basal melt'%(exp,year))
    #plt.tight_layout()
    
    return ax,mp
```

```python
fig,ax =plt.subplots()

ax,rp = plot_melt(ax,'oo0021',"2014-2018.wnetMean.nc",region=[-103,-99,-75.5,-74.2],levels=np.arange(0,101,1),figsize=None,ollie=True)

plt.show()
```
