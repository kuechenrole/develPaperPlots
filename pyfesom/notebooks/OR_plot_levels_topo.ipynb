```python
import sys
import os
sys.path.append("../")

import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
from netCDF4 import Dataset, MFDataset
import pyfesom as pf
import pandas as pd
```

```python
def plot_level_topo(meshpath,quant='draft',title='draft level depth'):

    mesh = pf.fesom_mesh(meshpath, abg=[0,0,0],cavity=True)

    draft=np.zeros_like(mesh.topo)
    for i,row in enumerate(mesh.n32):
        mesh.topo[i] = np.where(row>0)[0][-1]
        draft[i] = np.where(row>0)[0][0]

    if quant == 'draft':
        data = np.ma.masked_where(mesh.cflag!=1,draft)
    elif qunat == 'topo':
        data = mesh.topo
    elif qunat == 'wct':
        data = mesh.topo - mesh.draft

    plt.figure(figsize=(8,6))
    plt.scatter(mesh.x2,mesh.y2,40,data,cmap='viridis')
    cbar = plt.colorbar(orientation='horizontal', pad=0.03);
    plt.xlim(3.5,6.0)
    cbar.set_label('levels')
    plt.title(title)
    plt.tight_layout()
    plt.show()
    
def plot_depth_topo(meshpath,quant='shelf',title='shelf draft'):
    
    mesh = pf.fesom_mesh(meshpath, abg=[0,0,0],cavity=True)
    
    topoFile = os.path.join(meshpath,quant+'.out')
    file_content = pd.read_csv(topoFile, skiprows=0, names=['topo'] )
    topo  = file_content.topo.values

    plt.figure(figsize=(8,6))
    plt.scatter(mesh.x2,mesh.y2,40,topo,cmap='viridis')
    cbar = plt.colorbar(orientation='horizontal', pad=0.03);
    plt.xlim(3.5,6.0)
    cbar.set_label('depth in m')
    plt.title(title)
    plt.tight_layout()
    plt.show()
```
