```python
import sys
sys.path.append("../pyfesom")

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
%matplotlib notebook
from matplotlib import cm
from netCDF4 import Dataset

%load_ext autoreload
%autoreload 2

import pyfesom as pf
%aimport pyfesom

basedir = '/work/ollie/orichter/MisomipPlus/'
```

```python
rhofw = 1000
s2a = 3600*24*365.25
kg2gt = 10e-12

def get_melt_series(years):
    m = []
    for year in years:
        year = str(year)
        print('processing year '+year)
        meshpath  = basedir+'fesommesh/iceOcean1/'+year+'/'
        mesh = pf.fesom_mesh(meshpath, abg=[0,0,0],cavity=False)
        fl = Dataset(basedir+'fesomdata/iceOcean1/iceOcean1.'+year+'.forcing.diag.nc')
        
        mask = (mesh.cflag==1)
        melt = np.mean(fl['wnet'][:,mask],0)
        area = mesh.voltri[0:mesh.n2d][mask]
        
        m.append(np.sum(melt*area)*rhofw*s2a*kg2gt)
        
    return xr.DataArray(m,dims=('T'))
```

```python
ms = get_melt_series(range(1000,1201))
```
