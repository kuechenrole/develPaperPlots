```python
import sys
import os
sys.path.append("../pyfesom")

import pyfesom as pf
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
#%matplotlib notebook
%matplotlib inline
from matplotlib import cm
from netCDF4 import Dataset, MFDataset
import xarray as xr
import pdb
import cmocean.cm as ocm
```

```python
def make_bidx(mesh):
    bidx=np.ones(mesh.n2d,dtype=int)*-999
    for i,prof in enumerate(mesh.n32):
        ii = np.where(prof!=-999)
        bidx[i]=prof[ii[0][-1]]-1
    return bidx

def plot_bottomTS(ax,exp,year,month=False,region=False,
                  tlevels=np.arange(-2.2, 0, 0.1),slevels=np.arange(34.4, 35.5, 0.01),figsize=(9,11),ollie=False):
    
    #masterPath='/work/ollie/orichter/Misomip2/'
    if ollie:
        masterPath='/albedo/work/projects/oce_rio/orichter/ollieWork/Misomip2/'
        meshpath  =os.path.join(masterPath,exp,'fesomInitialMesh')
        dataPath=os.path.join(masterPath,exp,'fesomData','%s.%s.oce.mean.nc'%(exp,year))
    else:
        masterPath='/isibhv/netscratch/orichter/'
        meshpath  =os.path.join(masterPath,'timmsmesh',exp)
        dataPath=os.path.join(masterPath,'timmsdata',exp,'%s.%s.oce.mean.nc'%(exp,year))
        
    mesh = pf.fesom_mesh(meshpath, cavity=False)
    bidx=make_bidx(mesh)
    
    
    oce = xr.open_mfdataset(dataPath)
    mask = oce.temp[:,0]!=0
    if month==False:
        tbm = oce['temp'][mask][1:].mean('T').load()
       # sbm = oce['salt'][mask][1:].mean('T').load()
    else:
        tbm = oce['temp'][mask][month].load()
       # sbm = oce['salt'][mask][month].load()
    
    m=Basemap(projection='spstere',boundinglat=-60,lon_0=180)
    level_data, elem_no_nan = pf.get_data(tbm[bidx],mesh,0)

    #plt.figure(figsize=figsize)
    if region==False:
        m=Basemap(projection='spstere',boundinglat=-60,lon_0=180)
        m.drawmapboundary(fill_color='0.9')
        x, y = m(mesh.x2, mesh.y2)
    else:
        x, y = mesh.x2, mesh.y2
        ax.set_xlim([region[0],region[1]])
        ax.set_ylim([region[2],region[3]])
        
    if tlevels=='max':
        levels=np.arange(level_data.min(),level_data.max())
    else:
        levels = tlevels
        
    tp = ax.tricontourf(x, y, elem_no_nan[::], level_data, levels = levels, cmap=ocm.thermal, extend='both')
    #cbar = plt.colorbar(orientation='horizontal', pad=0.03);
    #cbar.set_label("Temperature, $^{\circ}$C")
    #plt.title('%s, %s, %s mean bottom temperature'%(exp,year,month))
    #plt.tight_layout()
    return ax,tp
    
def plot_melt(ax,exp,year,region=False,levels='max',figsize=(9,11),ollie=False):
    
    #masterPath='/work/ollie/orichter/Misomip2/'
    if ollie:
        masterPath='/albedo/work/projects/oce_rio/orichter/ollieWork/Misomip2/'
        meshpath  =os.path.join(masterPath,exp,'fesomInitialMesh')
        dataPath=os.path.join(masterPath,exp,'fesomData','%s.%s.forcing.diag.nc'%(exp,year))
    else:
        masterPath='/isibhv/netscratch/orichter/'
        meshpath  =os.path.join(masterPath,'timmsmesh',exp)
        dataPath=os.path.join(masterPath,'timmsdata',exp,'%s.%s.forcing.diag.nc'%(exp,year))
   
    mesh = pf.fesom_mesh(meshpath, cavity=False,get3d=True)
    frc = xr.open_mfdataset(dataPath)
    wnet = frc.wnet[1:].mean('T').where(mesh.cflag).load()*365.25*24*3600
    level_data, elem_no_nan = pf.get_data(wnet,mesh,0)
    
    #plt.figure(figsize=figsize)
    if region==False:
        m=Basemap(projection='spstere',boundinglat=-64,lon_0=180)
        m.drawmapboundary(fill_color='0.9')
        x, y = m(mesh.x2, mesh.y2)
        m.bluemarble(alpha=0.5)
    else:
        x, y = mesh.x2, mesh.y2
        ax.set_xlim([region[0],region[1]])
        ax.set_ylim([region[2],region[3]])

    if levels=='max':
        levels=np.arange(np.nanmin(level_data),np.nanmax(level_data))
    else:
        levels = levels

    mp = ax.tricontourf(x, y, elem_no_nan[::], level_data, levels = levels, cmap=ocm.thermal, extend='both')
    #cbar = plt.colorbar(orientation='horizontal');
    #cbar.set_label("ice shelf melt rate, m/yr")
    #plt.title('%s, %s mean basal melt'%(exp,year))
    #plt.tight_layout()
    
    return ax,mp
    
import matplotlib.colors as colors

from matplotlib.colors import LinearSegmentedColormap

cm_data = [[0.2081, 0.1663, 0.5292], [0.2116238095, 0.1897809524, 0.5776761905], 
 [0.212252381, 0.2137714286, 0.6269714286], [0.2081, 0.2386, 0.6770857143], 
 [0.1959047619, 0.2644571429, 0.7279], [0.1707285714, 0.2919380952, 
  0.779247619], [0.1252714286, 0.3242428571, 0.8302714286], 
 [0.0591333333, 0.3598333333, 0.8683333333], [0.0116952381, 0.3875095238, 
  0.8819571429], [0.0059571429, 0.4086142857, 0.8828428571], 
 [0.0165142857, 0.4266, 0.8786333333], [0.032852381, 0.4430428571, 
  0.8719571429], [0.0498142857, 0.4585714286, 0.8640571429], 
 [0.0629333333, 0.4736904762, 0.8554380952], [0.0722666667, 0.4886666667, 
  0.8467], [0.0779428571, 0.5039857143, 0.8383714286], 
 [0.079347619, 0.5200238095, 0.8311809524], [0.0749428571, 0.5375428571, 
  0.8262714286], [0.0640571429, 0.5569857143, 0.8239571429], 
 [0.0487714286, 0.5772238095, 0.8228285714], [0.0343428571, 0.5965809524, 
  0.819852381], [0.0265, 0.6137, 0.8135], [0.0238904762, 0.6286619048, 
  0.8037619048], [0.0230904762, 0.6417857143, 0.7912666667], 
 [0.0227714286, 0.6534857143, 0.7767571429], [0.0266619048, 0.6641952381, 
  0.7607190476], [0.0383714286, 0.6742714286, 0.743552381], 
 [0.0589714286, 0.6837571429, 0.7253857143], 
 [0.0843, 0.6928333333, 0.7061666667], [0.1132952381, 0.7015, 0.6858571429], 
 [0.1452714286, 0.7097571429, 0.6646285714], [0.1801333333, 0.7176571429, 
  0.6424333333], [0.2178285714, 0.7250428571, 0.6192619048], 
 [0.2586428571, 0.7317142857, 0.5954285714], [0.3021714286, 0.7376047619, 
  0.5711857143], [0.3481666667, 0.7424333333, 0.5472666667], 
 [0.3952571429, 0.7459, 0.5244428571], [0.4420095238, 0.7480809524, 
  0.5033142857], [0.4871238095, 0.7490619048, 0.4839761905], 
 [0.5300285714, 0.7491142857, 0.4661142857], [0.5708571429, 0.7485190476, 
  0.4493904762], [0.609852381, 0.7473142857, 0.4336857143], 
 [0.6473, 0.7456, 0.4188], [0.6834190476, 0.7434761905, 0.4044333333], 
 [0.7184095238, 0.7411333333, 0.3904761905], 
 [0.7524857143, 0.7384, 0.3768142857], [0.7858428571, 0.7355666667, 
  0.3632714286], [0.8185047619, 0.7327333333, 0.3497904762], 
 [0.8506571429, 0.7299, 0.3360285714], [0.8824333333, 0.7274333333, 0.3217], 
 [0.9139333333, 0.7257857143, 0.3062761905], [0.9449571429, 0.7261142857, 
  0.2886428571], [0.9738952381, 0.7313952381, 0.266647619], 
 [0.9937714286, 0.7454571429, 0.240347619], [0.9990428571, 0.7653142857, 
  0.2164142857], [0.9955333333, 0.7860571429, 0.196652381], 
 [0.988, 0.8066, 0.1793666667], [0.9788571429, 0.8271428571, 0.1633142857], 
 [0.9697, 0.8481380952, 0.147452381], [0.9625857143, 0.8705142857, 0.1309], 
 [0.9588714286, 0.8949, 0.1132428571], [0.9598238095, 0.9218333333, 
  0.0948380952], [0.9661, 0.9514428571, 0.0755333333], 
 [0.9763, 0.9831, 0.0538]]

parula_map = LinearSegmentedColormap.from_list('parula', cm_data)


def plot_resolution(ax,exp,year,region=False,levels='max',figsize=(9,11),ollie=False):
    
    #masterPath='/work/ollie/orichter/Misomip2/'
    if ollie:
        masterPath='/albedo/work/projects/oce_rio/orichter/ollieWork/Misomip2/'
        meshpath  =os.path.join(masterPath,exp,'fesomInitialMesh')
        dataPath=os.path.join(masterPath,exp,'fesomData','%s.%s.mesh.diag.nc'%(exp,year))
    else:
        masterPath='/isibhv/netscratch/orichter/'
        meshpath  =os.path.join(masterPath,'timmsmesh',exp)
        dataPath=os.path.join(masterPath,'timmsdata',exp,'%s.%s.mesh.diag.nc'%(exp,year))
    mesh = pf.fesom_mesh(meshpath, cavity=False,get3d=True)
    md = xr.open_mfdataset(dataPath)
    area = md.cluster_area.load()/1000/1000
    resolution=np.sqrt(area*4)*(3**(-0.25))
    level_data, elem_no_nan = pf.get_data(resolution,mesh,0)
    
    norm = colors.BoundaryNorm(boundaries=levels, ncolors=256)

    #plt.figure(figsize=figsize)
    if region==False:
        m=Basemap(projection='spstere',boundinglat=-64,lon_0=180)
        m.drawmapboundary(fill_color='0.9')
        x, y = m(mesh.x2, mesh.y2)
        m.bluemarble(alpha=0.5)
    else:
        x, y = mesh.x2, mesh.y2
        ax.set_xlim([region[0],region[1]])
        ax.set_ylim([region[2],region[3]])

    if levels=='max':
        levels=np.arange(np.nanmin(level_data),np.nanmax(level_data))
    else:
        levels = levels

    rp = ax.tricontourf(x, y, elem_no_nan[::], level_data/2, levels=levels[:-1], norm=norm, cmap=parula_map, extend='max')
    #cbar = plt.colorbar(plot,orientation='horizontal');
    #cbar.set_label("resolution, km")
    #plt.title('%s, %s mean basal melt'%(exp,year))
    #plt.tight_layout()
    return ax,rp
```
