```python
import pandas as pd
import matplotlib.dates as mdates

matplotlib.rcParams.update({'font.size': 14})

rhoi=916
kg2gt=10**(-12)
s2a=365.25*24*3600
regions={'FRIS':{'1':[-87.7,-45,-84,-74.6],'2':[-45,-23.0,-84,-77.34]},
         'PIG':{'1':[-103,-98,-75.6,-74.2]},
        'Totten':{'1':[113.0,118.0,-67.78,-66.35],'2':[118.0,119.65,-67.2,-66.35]}}

obs={'FRIS':{'Rignot':[0.32,0.10],'Adusumilli':[0.2,0.3]},
     'PIG':{'Rignot':[16.2,0.1],'Adusumilli':[14.0,1.6]},
     'Totten':{'Rignot':[8.15,0.6],'Adusumilli':[11.5,2.0]}}

def plot_bmb(wnet,area,region_key,ob_key,start='1/1/1979',title='',ylim=False,save=False):
    mask_region = np.zeros_like(mesh.x2)
    for region in regions[region_key].values():
        mask_region[(mesh.x2>region[0]) & (mesh.x2<region[1]) & (mesh.y2>region[2]) & (mesh.y2<region[3])]=1
    
    mask=np.zeros_like(mask_region)
    mask[(mask_region==1) & (mesh.cflag==1)]=1
    
    dA=area.where(mask)
    weights=dA/dA.sum()
        
    mm = (wnet*weights).sum('nodes_2d')*s2a
    bmb = (wnet*dA*rhoi).sum('nodes_2d')*s2a*kg2gt
    
    time= pd.date_range(start=start, periods=wnet.shape[0],freq='M')

    fig,ax=plt.subplots(figsize=(6,4))
    
    ax.plot(time,mm,'r',label='FESOM')
    ax.set_ylabel('mean melt rate in m/yr')
    if 'R' in ob_key:
        ob=obs[region_key]['Rignot']
        ax.hlines([ob[0]-ob[1],ob[0]+ob[1]],ax.get_xlim()[0],ax.get_xlim()[1],'k',label='Rignot 2013')
    if 'A' in ob_key:
        ob=obs[region_key]['Adusumilli']
        ax.hlines([ob[0]-ob[1],ob[0]+ob[1]],ax.get_xlim()[0],ax.get_xlim()[1],'b',label='Adusumilli 2020')
    
    #ax2=ax.twinx()
    #ax2.set_ylabel('basal mass loss in Gt/yr')
    #ax2.plot(time,bmb,'r')
    
    handles, _ = ax.get_legend_handles_labels()
    plt.legend(handles = handles)
    
    locator = mdates.YearLocator()
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(mdates.DateFormatter("\n%Y"))
    ax.xaxis.set_major_locator(plt.MaxNLocator(7))
    
    if ylim:
        ax.set_ylim(ylim)
    
    plt.title(title)
    if save:
        plt.savefig(save)
    plt.show()
```

```python
plot_bmb(fr.wnet[maskTime][:],meshDiag.cluster_area,'FRIS','RA',start='1/1/1979',title='',ylim=[-0.2,0.6],save='./figures/fesomIntSpinUpFRIS.png')
```
